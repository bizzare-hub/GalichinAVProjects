```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
def write_answer(ans, file_name):
    with open(file_name, "w") as fout:
        fout.write(str (ans))
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()
```

```python
digits_data = pd.DataFrame(digits.data)
digits_target = pd.DataFrame(digits.target)
breast_cancer_data = pd.DataFrame(breast_cancer.data)
breast_cancer_target = pd.DataFrame(breast_cancer.target)
```

```python
digits_data.head()
```

```python
breast_cancer_data.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
```

## Бернулли

```python
estimator = BernoulliNB()
score = cross_val_score(estimator, digits_data, np.array(digits_target).ravel(), scoring='accuracy', cv=3)
print ("accuracy по digits:")
print (score.mean())

print()

estimator = BernoulliNB()
score = cross_val_score(estimator, breast_cancer_data, np.array(breast_cancer_target).ravel(), scoring='accuracy', cv=3)
print ("accuracy по breast_cancer:")
print (score.mean())
```

## Мультиномиальное

```python
estimator = MultinomialNB()
score = cross_val_score(estimator, digits_data, np.array(digits_target).ravel(), scoring='accuracy', cv=3)
print ("accuracy по digits:")
print (score.mean())

print()

estimator = MultinomialNB()
score = cross_val_score(estimator, breast_cancer_data, np.array(breast_cancer_target).ravel(), scoring='accuracy', cv=3)
print ("accuracy по breast_cancer:")
print (score.mean())
```

## Гауссовское

```python
estimator = GaussianNB()
score = cross_val_score(estimator, digits_data, np.array(digits_target).ravel(), scoring='accuracy', cv=3)
print ("accuracy по digits:")
print (score.mean())

print()

estimator = GaussianNB()
score = cross_val_score(estimator, breast_cancer_data, np.array(breast_cancer_target).ravel(), scoring='accuracy', cv=3)
print ("accuracy по breast_cancer:")
print (score.mean())
```
